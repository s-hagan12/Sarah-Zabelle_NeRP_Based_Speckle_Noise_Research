```python
import numpy as np
from scipy.sparse import lil_matrix
import torch
import math
import cv2
import matplotlib.pyplot as plt
```

```python
#projections parameters
h=w=256 #height and width of the image
n_s = 256 #number of "bins" in each projection
num_proj = 20 #number of projections
```

```python
thetas = torch.linspace(0, np.pi, num_proj+1)[:-1]
print(thetas)
```

```python
R = lil_matrix((num_proj * n_s, h * w))
x = torch.arange(w, dtype=torch.float32) - w/2
y = torch.arange(w, dtype=torch.float32) - h/2
for theta_idx, theta in enumerate(thetas):
    r = [torch.cos(theta), torch.sin(theta)]
    scale = np.sqrt(np.dot(r,[w/2,h/2])**2)/(w/2)
    if theta > np.pi/2:
        scale = np.sqrt(np.dot(r,[-w/2,h/2])**2)/(w/2)
    #print(theta, scale)
    for i in x:
        for j in y:
            proj = np.dot(r,[i,j])/scale
            proj_bin = proj + n_s // 2
            lower = int(np.floor(proj_bin))
            l_weight = proj_bin - lower
            h_weight = 1 - l_weight
            
            pixel_idx = int((j + h/2) * w + (i + w/2))
            l_idx = theta_idx * n_s + lower
            
            if 0 <= l_idx < num_proj * n_s:
                R[l_idx, pixel_idx] += l_weight
            if 0 <= l_idx + 1 < num_proj * n_s:
                R[l_idx + 1, pixel_idx] += h_weight
```

```python
#Vectorise an image
def vec(x):
    r, c = x.shape
    v = np.zeros((r*c,1))
    for i in range(r):
        for j in range(c):
            v[i*c+j] = x[i][j]
    return v
#Convert from a vector back to the image matrix
def undo_vec(v, r, c):
    y = np.zeros((r,c))
    for i in range(r):
        for j in range(c):
            y[i][j] = v[i*c+j]
    return y
```

```python
v = vec(arr)
v = v.ravel() 
print(v.shape)
print(R.shape)
print(type(R))
```

```python
y = R.dot(v)
```

```python
proj_img = undo_vec(y, num_proj, n_s)
plt.imshow(proj_img, cmap='gray')
```

```python
from math import log10, sqrt

def psnr(img1, img2, max_pixel=255.0):
    mse = np.mean((img1 - img2) ** 2)
    if mse == 0:
        return float('inf')  # Images are identical
    return 20 * log10(max_pixel / sqrt(mse))
```
